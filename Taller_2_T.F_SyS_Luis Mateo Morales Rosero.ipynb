#1. Señal en Tiempo Continuo

Serie de Fourier (Exponencial, Trigonométrica y Compacta):

-Semejanzas: Las tres formas (exponencial, trigonométrica y compacta)
representan una señal periódica en tiempo continuo como una suma (infinita o finita para señales con un número finito de armónicos) de componentes sinusoidales o exponenciales complejos. Todas describen la señal en términos de sus frecuencias constituyentes y sus respectivas amplitudes y fases. Se aplican a señales periódicas.

**Diferencias:**

-Trigonométrica: Utiliza senos y cosenos en múltiplos enteros de la frecuencia fundamental de la señal. Produce coeficientes reales (a_n y b_n).

-Exponencial: Utiliza exponenciales complejos (e^(j n ω₀ t)). Produce coeficientes complejos (c_n) que contienen información de amplitud y fase. Es la forma más general y conveniente para el análisis teórico.

-Compacta: Combina los términos de seno y coseno de la forma trigonométrica en una sola sinusoide con una amplitud y fase específicas. Es útil para visualizar el espectro de amplitud y fase.

-Espectro: El espectro de la Serie de Fourier es discreto. Solo existen componentes de frecuencia en múltiplos enteros de la frecuencia fundamental (los armónicos). Cada línea en el espectro discreto corresponde a un armónico específico.

**Transformada de Fourier:**

-Semejanzas: También describe una señal en términos de sus frecuencias constituyentes.

-Diferencias: Se aplica a señales no periódicas en tiempo continuo. No hay una frecuencia fundamental única, por lo que la señal se descompone en un rango continuo de frecuencias.

-Espectro: El espectro de la Transformada de Fourier es continuo. Representa la densidad de energía o potencia de la señal en un rango continuo de frecuencias.

En resumen (Tiempo Continuo):

Señal Periódica (Serie de Fourier): Tiempo continuo -> Espectro discreto.
Señal No Periódica (Transformada de Fourier): Tiempo continuo -> Espectro continuo.

#2. Señal en Tiempo Discreto

Transformada de Fourier en Tiempo Discreto (DTFT):

-Semejanzas: También describe una señal en términos de sus frecuencias constituyentes. Relaciona una señal en el dominio del tiempo con una representación en el dominio de la frecuencia.

-Diferencias: Se aplica a señales no periódicas en tiempo discreto. La señal se representa como una secuencia de muestras discretas. El espectro resultante es continuo pero periódico (con periodo 2π en frecuencia angular normalizada). Esto se debe a la naturaleza muestreada de la señal en el tiempo.

-Espectro: El espectro de la DTFT es continuo y periódico.

**Transformada Discreta de Fourier (DFT):**

-Semejanzas: También describe una señal en términos de sus frecuencias constituyentes. Es una herramienta para analizar frecuencias en señales.

-Diferencias: Se aplica a señales periódicas o no periódicas de duración finita en tiempo discreto. En la práctica, a menudo se aplica a una secuencia finita de muestras de una señal, asumiendo que esta secuencia se repite periódicamente. Tanto la señal en el tiempo como el espectro son discretos. La DFT calcula el espectro en un número finito de puntos de frecuencia. Es la versión computacional de la Transformada de Fourier y la DTFT.

-Espectro: El espectro de la DFT es discreto. Consiste en un número finito de componentes de frecuencia.

#3. Algoritmo Fast Fourier Transform (FFT)

-¿En qué consiste el algoritmo FFT? El Fast Fourier Transform (FFT) no es una transformada diferente, sino un algoritmo computacional eficiente para calcular la Transformada Discreta de Fourier (DFT) y su inversa (IDFT). La DFT, calculada directamente, requiere un número significativo de operaciones (multiplicaciones y sumas) que puede ser computacionalmente costoso para secuencias largas. La FFT reduce drásticamente este costo aprovechando las propiedades de simetría y periodicidad de las operaciones matemáticas involucradas en la DFT.

-Utilidad para el cálculo de la Transformada Discreta de Fourier: La principal utilidad del FFT es hacer que el cálculo de la DFT sea prácticamente factible para señales largas. Antes del desarrollo de los algoritmos FFT, calcular la DFT de secuencias de longitud considerable era prohibitivamente lento. La FFT abrió la puerta a una amplia gama de aplicaciones en procesamiento de señales, análisis de datos, compresión, comunicaciones y muchas otras áreas donde el análisis de frecuencia es crucial.

-Detalle del Algoritmo FFT: La FFT más común es la de Cooley-Tukey, que se basa en la técnica de "divide y vencerás". La idea fundamental es descomponer el cálculo de una DFT de tamaño N en cálculos de DFT más pequeños. Esto se logra aprovechando la propiedad de que la DFT de una secuencia puede dividirse en la DFT de sus elementos pares y la DFT de sus elementos impares.

-Costo Computacional (DFT vs. FFT):

DFT Directa: Para calcular la DFT de una secuencia de longitud N directamente usando la fórmula, necesitas realizar N sumas, y cada suma involucra N multiplicaciones complejas. Por lo tanto, el número total de multiplicaciones complejas es aproximadamente N * N = N². El costo computacional se considera de orden O(N²).

FFT (Cooley-Tukey para N = 2^m): El algoritmo FFT reduce el costo computacional significativamente. Si N es una potencia de 2, el número de multiplicaciones complejas requeridas es aproximadamente (N/2) * log₂(N). El costo computacional se considera de orden O(N log N).

#Comparación del Costo Computacional:

La diferencia en el costo computacional es enorme, especialmente para secuencias largas.

-Para N = 1024 (una longitud común), la DFT directa requiere aproximadamente 1024² = 1,048,576 multiplicaciones. La FFT requiere aproximadamente (1024/2) * log₂(1024) = 512 * 10 = 5120 multiplicaciones. La FFT es miles de veces más rápida.

-Para N = 4096, la DFT directa es de orden 4096² ≈ 16.8 millones, mientras que la FFT es de orden 4096 * log₂(4096) ≈ 4096 * 12 ≈ 49,000.

Genere un ejemplo ilustrativo sobre Python en el que se grafique las senales en el tiempo y en frecuencia (utilizando la rfft) para: senal mensaje tipo pulso rectangular y senal mensaje tipo coseno. El usuario podra definir el ındice de modulacion de inter es. Ver cuaderno Modulacion AM

La **modulación por amplitud (AM)** con **detección coherente** es una técnica utilizada en telecomunicaciones para transmitir información a través de ondas de radio. A continuación, te explico en qué consiste y cuáles son sus aplicaciones principales.
 ¿En qué consiste la modulación por amplitud (AM)?

La modulación por amplitud consiste en **variar la amplitud de una onda portadora de alta frecuencia** en proporción a la señal de información (como voz o música) que se quiere transmitir. La frecuencia y fase de la portadora se mantienen constantes.

La señal modulada en AM se puede representar así:

$$
s(t) = [1 + m(t)] \cdot \cos(2\pi f_c t)
$$

Donde:

* $m(t)$ es la señal de mensaje (moduladora).
* $f_c$ es la frecuencia de la portadora.
* $\cos(2\pi f_c t)$ es la señal portadora.


### ¿Qué es la detección coherente?

La **detección coherente**, también llamada **demodulación síncrona**, es una técnica para recuperar la señal original $m(t)$ a partir de la señal AM recibida. Consiste en multiplicar la señal recibida por una copia sincronizada (en frecuencia y fase) de la portadora original.

Esto requiere que el receptor tenga un **oscilador local sincronizado** con el transmisor. Si hay desincronización, la recuperación de la señal puede fallar o degradarse.


### Aplicaciones de la modulación AM con detección coherente

* **Transmisión de radio AM**: Aunque en la práctica muchas emisoras usan detección no coherente (más simple), en ciertos casos (como comunicaciones militares o científicas) se prefiere detección coherente por su mejor desempeño frente al ruido.
* **Radiocomunicaciones aeronáuticas**: En algunas bandas de comunicación aérea se emplea AM coherente.
* **Modulación en cuadratura (QAM)**: Un tipo más avanzado de modulación que combina AM y detección coherente, muy usada en televisión digital, internet por cable y Wi-Fi.
* **Comunicaciones digitales**: Algunas formas de modulación digital como ASK (Amplitude Shift Keying) usan detección coherente para lograr mayor eficiencia espectral.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

# Parámetros
fs = 10000  # Frecuencia de muestreo (Hz)
t = np.linspace(0, 0.01, fs, endpoint=False)  # Tiempo
fc = 1000  # Frecuencia de la portadora (Hz)
Ac = 1     # Amplitud de la portadora

# Índice de modulación definido por el usuario
m_index = float(input("Ingrese el índice de modulación (ej. 0.5): "))

# Señales mensaje
# 1. Pulso rectangular
m_pulso = np.where((t >= 0.002) & (t <= 0.006), 1, 0)

# 2. Coseno de baja frecuencia
fm = 100  # Frecuencia del mensaje (Hz)
m_coseno = np.cos(2 * np.pi * fm * t)

# Modulación AM: s(t) = Ac * [1 + m(t) * m_index] * cos(2πfct)
s_pulso = Ac * (1 + m_index * m_pulso) * np.cos(2 * np.pi * fc * t)
s_coseno = Ac * (1 + m_index * m_coseno) * np.cos(2 * np.pi * fc * t)

# Espectros de frecuencia
f = rfftfreq(len(t), 1/fs)
S_pulso = np.abs(rfft(s_pulso))
S_coseno = np.abs(rfft(s_coseno))

# Gráficas
fig, axs = plt.subplots(4, 1, figsize=(10, 12))

axs[0].plot(t, s_pulso)
axs[0].set_title('Señal AM (mensaje: pulso rectangular) - Tiempo')
axs[0].set_xlabel('Tiempo [s]')
axs[0].set_ylabel('Amplitud')

axs[1].plot(f, S_pulso)
axs[1].set_title('Señal AM (pulso rectangular) - Frecuencia')
axs[1].set_xlabel('Frecuencia [Hz]')
axs[1].set_ylabel('Magnitud')

axs[2].plot(t, s_coseno)
axs[2].set_title('Señal AM (mensaje: coseno) - Tiempo')
axs[2].set_xlabel('Tiempo [s]')
axs[2].set_ylabel('Amplitud')

axs[3].plot(f, S_coseno)
axs[3].set_title('Señal AM (coseno) - Frecuencia')
axs[3].set_xlabel('Frecuencia [Hz]')
axs[3].set_ylabel('Magnitud')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq
from IPython.display import Audio

# Cargar audio
fs, audio = wavfile.read('fragmento_cancion.wav')  # asegúrate que es MONO

if audio.ndim == 2:
    audio = audio[:, 0]

audio = audio / np.max(np.abs(audio))  # Normalizar

# Duración para visualizar (ej. primeros 0.1 segundos)
dur = 10
N = int(fs * dur)
t = np.linspace(0, dur, N, endpoint=False)
mensaje = audio[:N]

# Parámetros de modulación
fc = 10000  # Frecuencia portadora en Hz
Ac = 1      # Amplitud portadora
m_index = 1 # Índice de modulación

# Portadora
portadora = Ac * np.cos(2 * np.pi * fc * t)

# Señal AM
s_am = Ac * (1 + m_index * mensaje) * np.cos(2 * np.pi * fc * t)

# Espectros
f = rfftfreq(N, 1/fs)
M_f = np.abs(rfft(mensaje))
P_f = np.abs(rfft(portadora))
AM_f = np.abs(rfft(s_am))

# Graficar señales en el tiempo
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(t, mensaje)
plt.title('Mensaje en el tiempo')
plt.xlabel('Tiempo [s]')

plt.subplot(3, 2, 2)
plt.plot(f, M_f)
plt.title('Espectro del mensaje')
plt.xlabel('Frecuencia [Hz]')

plt.subplot(3, 2, 3)
plt.plot(t, portadora)
plt.title('Portadora en el tiempo')
plt.xlabel('Tiempo [s]')

plt.subplot(3, 2, 4)
plt.plot(f, P_f)
plt.title('Espectro de la portadora')
plt.xlabel('Frecuencia [Hz]')

plt.subplot(3, 2, 5)
plt.plot(t, s_am)
plt.title('Señal AM en el tiempo')
plt.xlabel('Tiempo [s]')

plt.subplot(3, 2, 6)
plt.plot(f, AM_f)
plt.title('Espectro de la señal AM')
plt.xlabel('Frecuencia [Hz]')

plt.tight_layout()
plt.show()

# Reproducción de audio (mensaje, portadora, señal AM)
print("🔊 Reproduciendo señal mensaje (audio original):")
display(Audio(mensaje, rate=fs))

print("🔊 Reproduciendo portadora (sonido puro):")
display(Audio(portadora, rate=fs))

print("🔊 Reproduciendo señal AM:")
display(Audio(s_am, rate=fs))
```

Asumiendo $θ_0$ = 0, determine el espectro en cada una de
las etapas del sistema. Ademas, utilizando el fragmento de
la cancion de Youtube, grafique cada una de las etapas principales del de demodulacion en el tiempo y la frecuencia
(reproduzca el segmento de la cancion en cada etapa). Nota: para la etapa de filtrado pasa bajas, emplee filtrado ideal
espectral mediante FFT.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, irfft, rfftfreq
from IPython.display import Audio

# --- 1. Cargar mensaje (audio real) ---
fs, audio = wavfile.read("fragmento_cancion.wav")

# Si estéreo, tomar solo un canal
if audio.ndim == 2:
    audio = audio[:, 0]

# Normalizar
audio = audio / np.max(np.abs(audio))

# Fragmento a usar (ej. primeros 0.1 segundos)
dur = 10
N = int(fs * dur)
t = np.linspace(0, dur, N, endpoint=False)
mensaje = audio[:N]

# --- 2. Modulación AM ---
fc = 10000  # Frecuencia portadora
Ac = 1
m_index = 1

portadora = Ac * np.cos(2 * np.pi * fc * t)
s_am = Ac * (1 + m_index * mensaje) * portadora

# --- 3. Demodulación coherente (θ0 = 0) ---
demod = s_am * portadora  # Multiplicación coherente

# --- 4. Filtrado pasa bajas ideal por FFT ---
Demod_F = rfft(demod)
f = rfftfreq(N, 1/fs)
cutoff = 5000  # Hz

# Filtro ideal: cero todo lo mayor al corte
Demod_F_filtered = Demod_F.copy()
Demod_F_filtered[f > cutoff] = 0

mensaje_recuperado = irfft(Demod_F_filtered)

# --- 5. Función para graficar señal y espectro ---
def plot_signal_and_spectrum(signal, title, fs, N):
    f = rfftfreq(N, 1/fs)
    S = np.abs(rfft(signal))

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(t, signal)
    plt.title(f'{title} - Tiempo')
    plt.xlabel('Tiempo [s]')

    plt.subplot(1, 2, 2)
    plt.plot(f, S)
    plt.title(f'{title} - Frecuencia')
    plt.xlabel('Frecuencia [Hz]')
    plt.tight_layout()
    plt.show()

# --- 6. Graficar y reproducir cada etapa ---

# 1. Mensaje original
plot_signal_and_spectrum(mensaje, "1. Mensaje original", fs, N)
print("🔊 Reproduciendo mensaje original:")
display(Audio(mensaje, rate=fs))

# 2. Señal AM
plot_signal_and_spectrum(s_am, "2. Señal AM (modulada)", fs, N)
print("🔊 Reproduciendo señal AM:")
display(Audio(s_am, rate=fs))

# 3. Multiplicación coherente
plot_signal_and_spectrum(demod, "3. Señal multiplicada coherente", fs, N)
print("🔊 Reproduciendo señal demodulada (sin filtro):")
display(Audio(demod, rate=fs))

# 4. Después de filtrado (mensaje recuperado)
plot_signal_and_spectrum(mensaje_recuperado, "4. Mensaje recuperado (filtrado)", fs, N)
print("🔊 Reproduciendo mensaje recuperado:")
display(Audio(mensaje_recuperado, rate=fs))

# 5. Comparación directa
plt.figure(figsize=(10, 4))
plt.plot(t, mensaje, label='Mensaje original')
plt.plot(t, mensaje_recuperado, label='Mensaje recuperado', linestyle='--')
plt.title('Comparación: mensaje original vs recuperado')
plt.xlabel('Tiempo [s]')
plt.legend()
plt.tight_layout()
plt.show()
```
