#1. Señal en Tiempo Continuo

Serie de Fourier (Exponencial, Trigonométrica y Compacta):

-Semejanzas: Las tres formas (exponencial, trigonométrica y compacta)
representan una señal periódica en tiempo continuo como una suma (infinita o finita para señales con un número finito de armónicos) de componentes sinusoidales o exponenciales complejos. Todas describen la señal en términos de sus frecuencias constituyentes y sus respectivas amplitudes y fases. Se aplican a señales periódicas.

**Diferencias:**

-Trigonométrica: Utiliza senos y cosenos en múltiplos enteros de la frecuencia fundamental de la señal. Produce coeficientes reales (a_n y b_n).

-Exponencial: Utiliza exponenciales complejos (e^(j n ω₀ t)). Produce coeficientes complejos (c_n) que contienen información de amplitud y fase. Es la forma más general y conveniente para el análisis teórico.

-Compacta: Combina los términos de seno y coseno de la forma trigonométrica en una sola sinusoide con una amplitud y fase específicas. Es útil para visualizar el espectro de amplitud y fase.

-Espectro: El espectro de la Serie de Fourier es discreto. Solo existen componentes de frecuencia en múltiplos enteros de la frecuencia fundamental (los armónicos). Cada línea en el espectro discreto corresponde a un armónico específico.

**Transformada de Fourier:**

-Semejanzas: También describe una señal en términos de sus frecuencias constituyentes.

-Diferencias: Se aplica a señales no periódicas en tiempo continuo. No hay una frecuencia fundamental única, por lo que la señal se descompone en un rango continuo de frecuencias.

-Espectro: El espectro de la Transformada de Fourier es continuo. Representa la densidad de energía o potencia de la señal en un rango continuo de frecuencias.

En resumen (Tiempo Continuo):

Señal Periódica (Serie de Fourier): Tiempo continuo -> Espectro discreto.
Señal No Periódica (Transformada de Fourier): Tiempo continuo -> Espectro continuo.

#2. Señal en Tiempo Discreto

Transformada de Fourier en Tiempo Discreto (DTFT):

-Semejanzas: También describe una señal en términos de sus frecuencias constituyentes. Relaciona una señal en el dominio del tiempo con una representación en el dominio de la frecuencia.

-Diferencias: Se aplica a señales no periódicas en tiempo discreto. La señal se representa como una secuencia de muestras discretas. El espectro resultante es continuo pero periódico (con periodo 2π en frecuencia angular normalizada). Esto se debe a la naturaleza muestreada de la señal en el tiempo.

-Espectro: El espectro de la DTFT es continuo y periódico.

**Transformada Discreta de Fourier (DFT):**

-Semejanzas: También describe una señal en términos de sus frecuencias constituyentes. Es una herramienta para analizar frecuencias en señales.

-Diferencias: Se aplica a señales periódicas o no periódicas de duración finita en tiempo discreto. En la práctica, a menudo se aplica a una secuencia finita de muestras de una señal, asumiendo que esta secuencia se repite periódicamente. Tanto la señal en el tiempo como el espectro son discretos. La DFT calcula el espectro en un número finito de puntos de frecuencia. Es la versión computacional de la Transformada de Fourier y la DTFT.

-Espectro: El espectro de la DFT es discreto. Consiste en un número finito de componentes de frecuencia.

#3. Algoritmo Fast Fourier Transform (FFT)

-¿En qué consiste el algoritmo FFT? El Fast Fourier Transform (FFT) no es una transformada diferente, sino un algoritmo computacional eficiente para calcular la Transformada Discreta de Fourier (DFT) y su inversa (IDFT). La DFT, calculada directamente, requiere un número significativo de operaciones (multiplicaciones y sumas) que puede ser computacionalmente costoso para secuencias largas. La FFT reduce drásticamente este costo aprovechando las propiedades de simetría y periodicidad de las operaciones matemáticas involucradas en la DFT.

-Utilidad para el cálculo de la Transformada Discreta de Fourier: La principal utilidad del FFT es hacer que el cálculo de la DFT sea prácticamente factible para señales largas. Antes del desarrollo de los algoritmos FFT, calcular la DFT de secuencias de longitud considerable era prohibitivamente lento. La FFT abrió la puerta a una amplia gama de aplicaciones en procesamiento de señales, análisis de datos, compresión, comunicaciones y muchas otras áreas donde el análisis de frecuencia es crucial.

-Detalle del Algoritmo FFT: La FFT más común es la de Cooley-Tukey, que se basa en la técnica de "divide y vencerás". La idea fundamental es descomponer el cálculo de una DFT de tamaño N en cálculos de DFT más pequeños. Esto se logra aprovechando la propiedad de que la DFT de una secuencia puede dividirse en la DFT de sus elementos pares y la DFT de sus elementos impares.

-Costo Computacional (DFT vs. FFT):

DFT Directa: Para calcular la DFT de una secuencia de longitud N directamente usando la fórmula, necesitas realizar N sumas, y cada suma involucra N multiplicaciones complejas. Por lo tanto, el número total de multiplicaciones complejas es aproximadamente N * N = N². El costo computacional se considera de orden O(N²).

FFT (Cooley-Tukey para N = 2^m): El algoritmo FFT reduce el costo computacional significativamente. Si N es una potencia de 2, el número de multiplicaciones complejas requeridas es aproximadamente (N/2) * log₂(N). El costo computacional se considera de orden O(N log N).

#Comparación del Costo Computacional:

La diferencia en el costo computacional es enorme, especialmente para secuencias largas.

-Para N = 1024 (una longitud común), la DFT directa requiere aproximadamente 1024² = 1,048,576 multiplicaciones. La FFT requiere aproximadamente (1024/2) * log₂(1024) = 512 * 10 = 5120 multiplicaciones. La FFT es miles de veces más rápida.

-Para N = 4096, la DFT directa es de orden 4096² ≈ 16.8 millones, mientras que la FFT es de orden 4096 * log₂(4096) ≈ 4096 * 12 ≈ 49,000.
